```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 6))
sns.histplot(df['Price'], kde=True) # kde: 커널 밀도 추정
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Price') # 금액 분포도
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='PAYMENT TYPE', data=df)
plt.xlabel('Payment Type')
plt.ylabel('Count')
plt.ylim(0,2000)
plt.title('Count of Payment Types')
plt.show()
```

```python
numeric_df= df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='PAYMENT TYPE', y='Amount', data=df)
plt.xlabel('Payment Type')
plt.ylabel('Amount')
plt.title('Payment Type vs. Amount')
plt.show()
```

```python
sns.pairplot(df)
plt.title('Pairplot')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(df['Amount'])
plt.title('Boxplot of Amount')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df.set_index('Date', inplace=True)
monthly_sales = df.resample('M').sum()

plt.figure(figsize=(10,6))
plt.plot(monthly_sales['Amount'])
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Trend')
plt.show()
```
