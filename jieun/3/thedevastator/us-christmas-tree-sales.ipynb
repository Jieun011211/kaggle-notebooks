## 크리스마스 트리 판매
미국의 크리스마스 트리 판매

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/leejieun/kaggle-notebooks/datasets/thedevastator/us-christmas-tree-sales/Week 40 - US Christmas Tree Sales - 2010 to 2016.csv')
```

```python
df
```

```python
df.shape
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
sns.histplot(df['Average Tree Price'], kde=True)
plt.xlabel('Price')
plt.ylim(0,8)
plt.ylabel('frequency')
plt.show()
```

```python
df.dtypes
```

```python
df.drop('index', axis=1, inplace=True)
```

```python
numeric_df=df.select_dtypes(['float64','int64'])
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation_matrix')
plt.show()
```
