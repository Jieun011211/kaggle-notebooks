```python
import matplotlib.pyplot as plt

figure, axes = plt.subplots(ncols=2, nrows=1, figsize=(20,7))

axes[0].set_title('Price vs Rating')
sns.scatterplot(data= df, x='price', y='rating', ax=axes[0], alpha=.5)

axes[1].set_title('Rating vs Rating_count')
sns.scatterplot(data=df, x='rating', y='rating_count', ax=axes[1], alpha=.5)
```

```python
highest_avg_rating = df.groupby('merchant_id')['rating'].mean()
top_merchants = highest_avg_rating.sort_values(ascending=False)
print(f"Top 5 merchents are: {top_merchants.nlargest(5).index.tolist()}")

plt.xlim(1,6)
top_merchants = top_merchants.apply(lambda x: round(x))
sns.barplot(x=top_merchants.astype(int).index, y=df.groupby('merchant_id')['units_sold'].sum())
```
