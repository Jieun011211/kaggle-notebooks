## 영국 모기지 금리
영국 모기지 금리 시계열 날짜 (월간: 1999 - 2023)

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.ticker import FuncFormatter
sns.set_style('darkgrid')
```

```python
df= pd.read_csv('/Users/leejieun/kaggle-notebooks/datasets/liamhealy/uk-mortgage-rates/UK Mortgage Interest Rates.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df.set_index('Date', inplace=True)
```

```python
df.tail()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(df.index, df['Fixed_Rate_2y_95%_LTV'], label='Fixed 2y 95% LTV')
ax.plot(df.index, df['Fixed_Rate_2y_75%_LTV'], label='Fixed 2y 75% LTV')
ax.plot(df.index, df['Fixed_Rate_3y_75%_LTV'], label='Fixed 3y 75% LTV')
ax.plot(df.index, df['Tracker'], label='Tracker')
ax.plot(df.index, df['Variable Rate'], label='Variable')
ax.set(title='Mortgage Rates', xlabel='Date', ylabel='Rate')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _:'{:.1%}'.format(y)))
plt.show()
```
