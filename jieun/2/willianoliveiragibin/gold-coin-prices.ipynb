## 금화 가격
이 엑셀 파일에서, 이마미에 동전, 하프 토만, 그리고 쿼의 시계열 가격

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.animation as animation
```

```python
df = pd.read_csv('/home/leejieun/workspace/kaggle-notebooks/datasets/willianoliveiragibin/gold-coin-prices/1.csv')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(20,10))
plt.xlabel('Year')
plt.ylabel('High')

sns.lineplot(data=df, x='Date', y="High")
sns.lineplot(data=df, x='Date', y='Close')
plt.show()
```

```python
plt.figure(figsize=(40,20))
plt.xlabel('Price')
plt.ylabel('Weekday')
sns.stripplot(data=df, x='Weekday', y='Date')
plt.show()
```
