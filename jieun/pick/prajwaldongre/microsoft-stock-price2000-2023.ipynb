```python
from plotly.express import line
import datetime
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
line(data_frame=df, x='Date', y=['Close', 'Adj Close'])
```

```python
line(data_frame=df, x='Date',  y=['Volume'],log_y=True) #log_y: 눈금 간격이 10
```

```python
from plotly.express import scatter
scatter(data_frame=df, y='Volume', x='Adj Close', color=df['Date'].dt.year, log_y=True, height=1000)
```

```python
scatter(data_frame=df, x='Date', y=['Volume'], log_y=True, trendline='ols', trendline_color_override='orange') # trendline : 추세선(ordinary least square의 약자로 최소자승법으로 회귀 모델)
```

```python
import plotly.express as px
```

```python
fig = px.histogram(df, x='Volume')

fig.update_xaxes(
    rangeslider_visible=True,
)

fig.show()
```
