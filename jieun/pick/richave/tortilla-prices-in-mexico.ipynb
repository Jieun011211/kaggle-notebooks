## 멕시코의 토르티야 가격
2007년부터 2024년까지 멕시코에서 가장 관련성이 높은 식품("토르티야")의 가격

```python
import pandas as pd
```

```python
df = pd.read_csv("/home/leejieun/workspace/kaggle-notebooks/datasets/richave/tortilla-prices-in-mexico/tortilla_prices.csv")
df.head()
```

```python
print(df.isna().sum(), df.shape)
```

```python
ser = df["Price per kilogram"].isna()
na_indices = ser.index[ser == True]
df_clean = df.drop(na_indices)
```

```python
df_clean['State'].value_counts().keys().shape
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df_state = pd.DataFrame(df_clean.groupby(["State", "Year"])["Price per kilogram"].mean())
df_state.head()
```

```python
df_state_bcr = df_state.reset_index().pivot(index="Year", columns="State", values="Price per kilogram")
df_state_bcr
```
