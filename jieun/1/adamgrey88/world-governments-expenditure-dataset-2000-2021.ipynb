```python
def ulkeSec(ulke):
    df=pd.read_csv('/home/leejieun/workspace/kaggle-notebooks/datasets/adamgrey88/world-governments-expenditure-dataset-2000-2021/WorldExpenditures.csv')
    ulkedf=df[df['Country'] ==ulke]
    return ulkedf
```

```python
def harcamaPlot(ulke, sektor):
    tr=ulkeSec(ulke)
    egitim = tr[tr['Sector'] ==sektor]
    egitim.head(3)
    plt.style.use("fivethirtyeight")
    plt.figure(figsize=(16,5))
    plt.xlabel('Year')
    plt.ylabel('Expenditure(million USD)')
    plt.plot(egitim['Year'].astype('str'), egitim['Expenditure(million USD)'])
    plt.show()
```

```python
harcamaPlot("Türkiye", "Education")
```

```python
harcamaPlot("Türkiye","Health")
```

```python
harcamaPlot('Türkiye','Defence')
```
