```python
df['Date '] = pd.to_datetime(df['Date '], format='%d-%b-%Y')
```

```python
df
```

```python
new_date_format =[]
for nif in df['Date ']:
    year,month,day = str(nif.date()).split('-')
    new_date_str = "-".join([day, month, year])
    new_date_format.append(new_date_str)
```

```python
df['Date '] = new_date_format
```
